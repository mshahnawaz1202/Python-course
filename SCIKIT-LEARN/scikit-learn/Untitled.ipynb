```python
import pandas as pd

data={
    'Name':['Ali','Haider','Hamza','Bilal','Shahnawaz'],
    'Age':[20,None,34,None,19],
    'Salary':[12000,45000,None,23000,105000]
}
df=pd.DataFrame(data)
print('\n------------------------------')
print(df)
print('\n------------------------------')
print(df.isnull().sum())
print('\n------------------------------')
print(df.isnull().mean()*100)
print('\n------------------------------')
df_drop=df.dropna()
print(df_drop)
print('\n------------------------------')
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
print(df)
print('\n------------------------------')




```
