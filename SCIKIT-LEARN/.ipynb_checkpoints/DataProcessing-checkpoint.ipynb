```python
'''
in this we have to make all values in one range like close to 0
height         study
180             7
170             10
160              3

height is in 100s and study is in 10s so we have to bring all in one range so that ml model can easily learn form them



for this we used standard scaler
it bring all values so that their mean is 0 and standard deviation is 1
mostly best for normal distribution where there is bell shaped data

min/mx scalar if we have to scale between 0 and 1 
used when we donot want -ve values

'''



# from sklearn.preprocessing import StandardScaler,MinMaxScaler
# scalar=StandardScaler()

# X_Scaled=scalar.fit_transform()

# scalar=MinMaxScaler()

# x_sacaled=scalar.fit_transform()
# --------------------------------------------------------------------------------------------------------------
# x-> actual value 
# Xmin -> minnimum value in column
# x max-> max cvalue i columns

import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
data={
    'Hours':[1,2,3,4,5],
    'TestScores':[60,70,75,80,85]
}

df=pd.DataFrame(data)

stand_sca=StandardScaler()
stand_scal=stand_sca.fit_transform(df)
#  z=(actual_value-mean)/standard_deviation

print('\n------------- Standard Scalar Output ---------------\n')

print(pd.DataFrame(stand_scal,columns=['Hours','TestScores']))





min_max_sca=MinMaxScaler()
min_max_scaled=min_max_sca.fit_transform(df)
#  scaled= (X-Xmin)/(Xmax-Xmin)


print('\n--------- MinMax Scalar Output ---------\n')

print(pd.DataFrame(min_max_scaled,columns=['Hours','TestScores']))

X= df[['Hours']]
Y=df[['TestScores']]

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
'''
here is splitting of data
x_train and y_train will use 80% data for training
x_test and y_test will use 20% of data for testing
'''

print('\n\nTraining Data')

print(X_train)

print('\nTest Data')
print(X_test)



print('\n--------- Test Scores -----------')

print('\n\nTraining Data')

print(Y_train)

print('\nTest Data')
print(Y_test)

```

```python
# --------------------------- Logistic Regression ---------------------
# for predicting label data
# range : 0-1
# 0->No
# 1-> Yes


from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
X= [[1],[2],[3],[4],[5]] # hrs study
Y=[0,0,1,1,1] # pass/fail
model.fit(X,Y)
hours=float(input('Enter how many hours you studied : '))
predicted_data=model.predict([[hours]])[0]

if predicted_data==1:
    print(f"You study ({hours}) Hours.You will Likely to Be Pass.")
else:
    print(f"You study ({hours}) Hours.You will Likely to Be Fail.")






```

```python
# ----------------------------- KNN -------------------
from sklearn.neighbors import KNeighborsClassifier
X=[
    # [Rating,Duration]
    [8,160],
    [6.2,170],
    [7.2,168],
    [8.2,155]
]
# Y -> Genre
# 0-> Action Movie
# 1 -> Comedy
Y = [0,0,1,1] 

model= KNeighborsClassifier(n_neighbors=3)

model.fit(X,Y)

rating=float(input('Enter Rating : '))
Duration=float(input('Enter Duration of Movie : '))
predicted_data=model.predict([[rating,Duration]])[0]

if predicted_data==1:
    print(f"Genre for Movie with Duration ({Duration}) and Rating({rating}) is Comedy")
else:
    print(f"Genre for Movie with Duration ({Duration}) and Rating({rating}) is Action")


```

```python
# -------------------- Decision Tree -----------------
from sklearn.tree import DecisionTreeClassifier
# 1 -> pure apple
# 2-> pure orange
# for colors
X=[
    # [size/weoight,color]
    [7,2], 
    [8,3],
    [9,8],
    [10,9]
]

#fruits
# 0->Apple
#1->Orange
Y=[0,0,1,1]

model= DecisionTreeClassifier()

model.fit(X,Y)

size=float(input('Enter size of fruit : '))
shade=float(input('Enter Shade of fruit (1-10) : '))
predicted_data=model.predict([[size,shade]])[0]

if predicted_data==1:
    print('This is likely Orange')
else:
    print('This is likely Apple')

```

```python
# ---------------------------- KMean Classification ------------------------------
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# data={
#     'Customer':['Shahnawaz','Ali hassan','Haider','Umair','Huraira'],
#     'Age':[19,20,21,21,18],
#     'Spending':[1000,2000,5000,2500,6000]
# }


data={
    'Sr.No.':['C1','C2','C3','C4','C5','C6','C7'],
    'Age':[20,40,30,18,28,35,45],
    'Amount':[500,1000,800,300,1200,1400,1800]
}

df=pd.DataFrame(data)
# X= df[['Age','Spending']]
X = df[['Age','Amount']]
model=KMeans(n_clusters=3,random_state=42,n_init=10)
df['Group']=model.fit_predict(X)
plt.figure(figsize=(6,5))
for group in df['Group'].unique():
    group_data=df[df['Group']==group]
    # plt.scatter(group_data['Age'],group_data['Spending'],label=f'Group{group}')
    plt.scatter(group_data['Age'],group_data['Amount'],label=f'Group{group}')

plt.xlabel('AGE')
plt.ylabel('Amount')
plt.grid(True)
plt.title('Clusters (K-Mean)')
plt.legend()
plt.show()

print('\n\n')
print(df)
```

```python
# ---------------------------- K-Medoids Clustering ----------------------------
import pandas as pd
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids

# ---------------------------- Dataset ----------------------------
data = {
    'Customer': ['Shahnawaz', 'Ali Hassan', 'Haider', 'Umair', 'Huraira'],
    'Age': [19, 20, 21, 21, 18],
    'Spending': [1000, 2000, 5000, 2500, 6000]
}

df = pd.DataFrame(data)

# Features
X = df[['Age', 'Spending']]

# ---------------------------- K-Medoids Model ----------------------------
model = KMedoids(
    n_clusters=2,
    metric='euclidean',
    init='k-medoids++',
    random_state=42
)

# Fit & Predict
df['Group'] = model.fit_predict(X)

# ---------------------------- Output ----------------------------
print(df)

# ---------------------------- Visualization ----------------------------
plt.figure(figsize=(6, 5))

for group in df['Group'].unique():
    g = df[df['Group'] == group]
    plt.scatter(g['Age'], g['Spending'], label=f'Group {group}')

# Plot medoids
medoids = model.cluster_centers_
plt.scatter(medoids[:,0], medoids[:,1], marker='X', s=200)

plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('Customer Segmentation using K-Medoids (sklearn)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -------------------- Elbow Method ----------------------
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dataset
data = {
    'Age': [19, 20, 21, 21, 18],
    'Spending': [1000, 2000, 5000, 2500, 6000]
}

df = pd.DataFrame(data)
X = df[['Age', 'Spending']]

# Elbow Method
wcss = []

for k in range(1, 6):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X)
    wcss.append(model.inertia_)

# Plot
plt.figure(figsize=(6,5))
plt.plot(range(1,6), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for K-Means')
plt.grid(True)
plt.show()
```
